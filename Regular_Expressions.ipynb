```python
import pandas as pd
import re
```

```python
text_data = [
    "This is a sample text with numbers like 123 and special characters like @#$",
    "Cleaning text is crucial for effective Natural Language Processing",
    "Lets explore how regular expressions can help in text cleaning"
]
```

```python
type(text_data)
```

```python
df = pd.DataFrame({"text" :text_data})
```

```python
df
```

```python
df["text"][0]
```

```python
df["Cleaned_Text"] = df["text"].apply(lambda x :re.sub(r"[^A-Za-z0-9/s]"," ",x))
```

```python
df["text"][0]
```

```python
df["Cleaned_Text"][0]
```

```python
df
```

```python
df["Cleaned_Text"] = df["Cleaned_Text"].apply(lambda x : x.lower())
```

```python
df
```

```python
df["Cleaned_Text"] = df["Cleaned_Text"].apply(lambda x : re.sub(r"\d"," ",x))
```

```python
df
```

```python
df["Cleaned_Text"] = df["Cleaned_Text"].apply(lambda x : re.sub(r"[^\w\s]"," ",x))
```

```python
df
```

```python
df["Cleaned_Text"] = df["Cleaned_Text"].apply(lambda x : re.sub(r"\s+"," ",x).strip())
```

```python
df
```

```python
stop_word = ["this", "is", "and", "for", "in", "which"]
```

```python
df["Cleaned_Text"] = df["Cleaned_Text"].apply(lambda x : " ".join([word for word in x.split()if word.lower() not in stop_word]))
```
