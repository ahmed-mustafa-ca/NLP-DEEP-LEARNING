```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
sentence = "Natural Language Processing is amazing!"
```

```python
words = word_tokenize(sentence)
```

```python
words
```

```python
pos_tags = pos_tag(words)
```

```python
pos_tags
```

```python
import pandas as pd
```

```python
data = {"sentences" : ["NLTK is powerful tool for NLP",
                       "It provides various functionalites",
                       "Turkey is a great country"]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df["POS_taggings"] = df["sentences"].apply(lambda x : pos_tag(word_tokenize(x)))
```
