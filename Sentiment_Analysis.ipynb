```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
sample_sentence = "I hated this city "
#"I love this meal,it was fantastic,ver amazing"
 #  
#"I absolutely loved the movie ,it was fantastic"
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
sentiment_scores = sid.polarity_scores(sample_sentence)
```

```python
sentiment_scores
```

```python
import pandas as pd
```

```python
data = {"sentences" : ["This product is terrible!",
                       "I am never coming back again, it was horrible",
                       "I loved the food",
                       "I love being in this city",
                       "Such a wonderful place"]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df["Sentiment Scores"] = df["sentences"].apply(lambda x :sid.polarity_scores(x)["compound"])
```
