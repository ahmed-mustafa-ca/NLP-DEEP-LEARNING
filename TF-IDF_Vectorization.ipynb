```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
sentence = "Natural Language Processing is a fascinating feild to study. Natural Language Processing is great. So i like to study great things. is and is "
```

```python
tfidf_vectorizer = TfidfVectorizer()
```

```python
tfidf_matrix = tfidf_vectorizer.fit_transform([sentence])
```

```python
tfidf_matrix
```

```python
tfidf_matrix.toarray()
```

```python
pd.DataFrame(tfidf_matrix.toarray(), columns= tfidf_vectorizer.get_feature_names_out())
```

```python
data = {"text": ["NLP is great",
                 "Python is amazing studying",
                 "I love studying Python",
                 "I love studying NLP"]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
tfidf_matrix_df = tfidf_vectorizer.fit_transform(df["text"])
```
