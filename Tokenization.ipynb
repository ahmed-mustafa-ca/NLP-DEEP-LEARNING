```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
dataset = [
    "Tokenization is an essential part in NLP",
    "Stemming reduces words to their root form",
    "Natural Language processing is great!"
]
```

```python
tokenized_sentences = [word_tokenize(sentence)for sentence in dataset]
```

```python
tokenized_sentences
```

```python
porter = PorterStemmer()
```

```python
stemmed_sentences = [[porter.stem(word)for word in sentence]for sentence in tokenized_sentences]
```

```python
stemmed_sentences
```
